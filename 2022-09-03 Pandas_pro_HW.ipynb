#### Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

* если источник traffic_source равен yandex или google, то в source_type ставится organic
* для источников paid и email из России - ставим ad
* для источников paid и email не из России - ставим other
* все остальные варианты берем из traffic_source без изменений

```python
log = pd.read_csv('visit_log.csv', sep=';')
# log.head()
```

```python
def traffic(row):
    if 'yandex' in row.traffic_source:
        return 'organic'
    if 'google' in row.traffic_source:
        return 'organic'
    if 'paid' in row.traffic_source and 'Russia' in row.region :
        return 'ad'
    if 'mail' in row.traffic_source and 'Russia' in row.region :
        return 'ad'
    if 'paid' in row.traffic_source and 'Russia' not in row.region :
        return 'other'
    if 'mail' in row.traffic_source and 'Russia' not in row.region :
        return 'other'
    else:
        return row.traffic_source
```

```python
log['source_type'] = log.apply(traffic, axis=1)
log.head()
```

#### Задание 3

1. Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, 
   которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца    timestamp для данного значения userId.

```python
rating = pd.read_csv('ratings.csv')
rating.head()
```

```python
rating.groupby(['userId', 'rating']).agg({'timestamp': 'sum'})
```

```python
def min_max(column):
    """
    Ср.время жизни в часах
    
    """
    return (column.max() - column.min()) / (60 * 60)
```

```python
av_life_user = rating.groupby(['userId']).agg({'rating': 'count', 'timestamp': min_max})
av_life_user[(av_life_user['rating'] > 100)].sort_values(by=['timestamp']).reset_index()
```

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

Дана статистика услуг перевозок клиентов компании по типам:

- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
cons_tab = (rzd.merge(auto, how='outer', on='client_id')).merge(air, how='outer', on='client_id')
cons_tab.fillna(0)
```

```python
client_base_cons = client_base.merge(cons_tab, how='left', on='client_id')
client_base_cons.fillna(0)
```
